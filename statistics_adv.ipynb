```python
# 1.Q= Generate a list of 100 integers containing values between 90 to 130 and store it in the variable int_list. 
# After generating the list, find the following:
# (i) Write a Python function to calculate the mean of a given list of numbers.Create a function to find the median 
# of a list of numbers.
# (ii) Develop a program to compute the mode of a list of integers.
# (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
# (iv) Write a Python function to find the geometric mean of a list of positive numbers.
# (v) Create a program to calculate the harmonic mean of a list of values.
# (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
# (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of 
# outliers.

import random
from collections import Counter
from math import prod

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Mean and Median Functions
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    numbers.sort()
    n = len(numbers)
    mid = n // 2
    if n % 2 == 0:
        return (numbers[mid - 1] + numbers[mid]) / 2
    else:
        return numbers[mid]

# (ii) Mode Function
def calculate_mode(numbers):
    
    frequency = Counter(numbers)
    max_freq = max(frequency.values())
    modes = [key for key, value in frequency.items() if value == max_freq]
    return modes

# (iii) Weighted Mean Function
def calculate_weighted_mean(values, weights):
    total_weight = sum(weights)
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    return weighted_sum / total_weight

# (iv) Geometric Mean Function
def calculate_geometric_mean(numbers):
    return prod(numbers) ** (1 / len(numbers))

# (v) Harmonic Mean Function
def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1 / x for x in numbers)

# (vi) Midrange Function
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (vii) Trimmed Mean Function
def calculate_trimmed_mean(numbers, trim_percent):
    n = len(numbers)
    trim_count = int(n * trim_percent / 100)
    numbers.sort()
    trimmed_numbers = numbers[trim_count:-trim_count] if trim_count > 0 else numbers
    return calculate_mean(trimmed_numbers)

print("Generated List:", int_list)
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
weights = [random.randint(1, 10) for _ in range(100)]  # Generate random weights for weighted mean
print("Weighted Mean:", calculate_weighted_mean(int_list, weights))
print("Geometric Mean:", calculate_geometric_mean(int_list))
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))

# output:
# Generated List: [103, 104, 107, 121, 96, 126, 123, 105, 114, 102, 120, 120, 125, 126, 101, 106, 96, 127, 101,92, 125, 104, 119, 127, 104, 128, 91, 128, 102, 95, 117, 127, 109, 109, 127, 109, 95, 118, 115, 130, 90, 96, 92, 103, 103, 124, 105, 121, 109, 114, 127, 127, 123, 125, 121, 116, 104, 125, 104, 118, 115, 100, 108, 127, 130, 115, 121, 124, 99, 126, 130, 94, 127, 114, 107, 100, 103, 108, 103, 91, 119, 123, 124, 97, 103, 120, 112, 123, 121, 92, 129, 105, 120, 91, 107, 106, 122, 129, 124, 100]
# Mean: 112.5
# Median: 114.0
# Mode: [127]
# Weighted Mean: 112.6145648312611
# Geometric Mean: 111.85077837191129
# Harmonic Mean: 111.18982957641572
# Midrange: 110.0
# Trimmed Mean (10%): 113.025
```

```python
# 3. Write a Python class representing a discrete random variable with methods to calculate its expected 
# value and variance.

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not all(0 <= p <= 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")
        if abs(sum(probabilities) - 1) > 1e-6:
            raise ValueError("The sum of probabilities must equal 1.")
        self.values = values
        self.probabilities = probabilities
    
    def expected_value(self):
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))
    
    def variance(self):
        mean = self.expected_value()
        return sum(prob * (value - mean) ** 2 for value, prob in zip(self.values, self.probabilities))

values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]
random_variable = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", random_variable.expected_value())
print("Variance:", random_variable.variance())

# output:
# Expected Value: 3.2
# Variance: 1.56
```

```python
# 4.Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and 
# variance of the outcomes.

import random

class FairDie:
    def __init__(self, sides=6):
        self.sides = sides
        self.values = list(range(1, sides + 1))
        self.probabilities = [1 / sides] * sides 
    def roll(self):
        return random.choice(self.values)

    def expected_value(self):
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum(prob * (value - mean) ** 2 for value, prob in zip(self.values, self.probabilities))

fair_die = FairDie()
rolls = [fair_die.roll() for _ in range(1000)]
expected_value = fair_die.expected_value()
variance = fair_die.variance()

print("Simulated Rolls:", rolls[:10], "...")  
print("Expected Value (Theoretical):", expected_value)
print("Variance (Theoretical):", variance)

# Calculate the sample mean and variance from simulated rolls
sample_mean = sum(rolls) / len(rolls)
sample_variance = sum((x - sample_mean) ** 2 for x in rolls) / len(rolls)

print("Sample Mean (Simulated):", sample_mean)
print("Sample Variance (Simulated):", sample_variance)

# output:
# Simulated Rolls: [5, 4, 3, 4, 2, 5, 3, 1, 1, 6] ...
# Expected Value (Theoretical): 3.5
# Variance (Theoretical): 2.9166666666666665
# Sample Mean (Simulated): 3.613
# Sample Variance (Simulated): 2.8832310000000194
```

```python
# 5. Create a Python function to generate random samples from a given probability distribution (e.g., 
# binomial, Poisson) and calculate their mean and variance.

import numpy as np

def generate_samples(distribution, params, sample_size=1000):
    if distribution == 'binomial':
        if 'n' not in params or 'p' not in params:
            raise ValueError("Parameters for binomial must include 'n' and 'p'.")
        n, p = params['n'], params['p']
        samples = np.random.binomial(n, p, sample_size)
    elif distribution == 'poisson':
        if 'lambda' not in params:
            raise ValueError("Parameter for poisson must include 'lambda'.")
        lam = params['lambda']
        samples = np.random.poisson(lam, sample_size)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")
    
    mean = np.mean(samples)
    variance = np.var(samples)
    return samples, mean, variance

# Binomial distribution: n=10, p=0.5
samples_binomial, mean_binomial, var_binomial = generate_samples(
    'binomial', {'n': 10, 'p': 0.5}, sample_size=1000
)
print("Binomial Distribution:")
print("Mean (Simulated):", mean_binomial)
print("Variance (Simulated):", var_binomial)

# Poisson distribution: lambda=4
samples_poisson, mean_poisson, var_poisson = generate_samples(
    'poisson', {'lambda': 4}, sample_size=1000
)
print("\nPoisson Distribution:")
print("Mean (Simulated):", mean_poisson)
print("Variance (Simulated):", var_poisson)

# output:
# Binomial Distribution:
# Mean (Simulated): 4.925
# Variance (Simulated): 2.373375

# Poisson Distribution:
# Mean (Simulated): 4.006
# Variance (Simulated): 3.939964
```

```python
# 6.Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute 
# the mean, variance, and standard deviation of the samples.

import numpy as np

def gaussian_distribution_statistics(mean, std_dev, sample_size=1000):
    samples = np.random.normal(loc=mean, scale=std_dev, size=sample_size)
    
    calculated_mean = np.mean(samples)
    calculated_variance = np.var(samples)
    calculated_std_dev = np.std(samples)
    
    return samples, calculated_mean, calculated_variance, calculated_std_dev

# Example usage
mean = 15  # Mean of the distribution
std_dev = 5  # Standard deviation of the distribution
sample_size = 1000  # Number of samples

samples, calc_mean, calc_variance, calc_std_dev = gaussian_distribution_statistics(mean, std_dev, sample_size)

print("First 10 Generated Samples:", samples[:10])
print(f"Simulated Mean: {calc_mean:.2f}")
print(f"Simulated Variance: {calc_variance:.2f}")
print(f"Simulated Standard Deviation: {calc_std_dev:.2f}")

# output:
# First 10 Generated Samples: [15.92385441 11.802931   12.16423747 17.75314688 29.80998268 16.48384285
#   8.35610973 10.81125033 12.89372413 13.11483675]
# Simulated Mean: 15.04
# Simulated Variance: 24.38
# Simulated Standard Deviation: 4.94
```

```python
# 8.Write a Python function to calculate the probability density function (PDF) of a continuous random 
# variable for a given normal distribution.

import math

def normal_pdf(x, mean, std_dev):
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = math.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))
    return coefficient * exponent

# Example usage
mean = 0  # Mean of the normal distribution
std_dev = 1  # Standard deviation of the normal distribution

# Evaluate the PDF at various points
points = [-2, -1, 0, 1, 2]
pdf_values = [normal_pdf(x, mean, std_dev) for x in points]

print("Point\tPDF Value")
for point, pdf_value in zip(points, pdf_values):
    print(f"{point}\t{pdf_value:.6f}")

# output:
# Point	PDF Value
# -2	0.053991
# -1	0.241971
# 0	0.398942
# 1	0.241971
# 2	0.053991
```

```python
# 9.Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

import math

def exponential_cdf(x, lambda_):
    if x < 0:
        return 0.0
    return 1 - math.exp(-lambda_ * x)

# Example usage
lambda_ = 0.5  # Rate parameter of the exponential distribution
points = [0, 1, 2, 3, 4, 5]  # Points at which to evaluate the CDF

print("Point\tCDF Value")
for point in points:
    cdf_value = exponential_cdf(point, lambda_)
    print(f"{point}\t{cdf_value:.6f}")

# output:
# Point	CDF Value
# 0	0.000000
# 1	0.393469
# 2	0.632121
# 3	0.776870
# 4	0.864665
# 5	0.917915
```

```python
# 10.Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

import math

def poisson_pmf(k, lambda_):
    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer.")
    if lambda_ <= 0:
        raise ValueError("lambda_ must be greater than 0.")
    return (math.exp(-lambda_) * (lambda_ ** k)) / math.factorial(k)

# Example usage
lambda_ = 4  # Rate parameter of the Poisson distribution
values = [0, 1, 2, 3, 4, 5]  # Points at which to evaluate the PMF

print("k\tPMF Value")
for k in values:
    pmf_value = poisson_pmf(k, lambda_)
    print(f"{k}\t{pmf_value:.6f}")

# output:
# k	PMF Value
# 0	0.018316
# 1	0.073263
# 2	0.146525
# 3	0.195367
# 4	0.195367
# 5	0.156293
```

```python
# 11.A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors 
# who make a purchase). They collect data from the old and new layouts to compare.
# To generate the data use the following command:
# python
import numpy as np
# 50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors  
new_layout = np.array([1] * 70 + [0] * 930)
# Apply z-test to find which layout is successful.

from statsmodels.stats.proportion import proportions_ztest
# Number of successes and total observations
successes = [old_layout.sum(), new_layout.sum()]  # Total purchases
n_obs = [len(old_layout), len(new_layout)]       # Total visitors

# Perform the Z-test
z_stat, p_value = proportions_ztest(successes, n_obs, alternative='larger')

print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("The new layout has a significantly higher conversion rate (reject null hypothesis).")
else:
    print("No significant difference in conversion rates (fail to reject null hypothesis).")

# output:
# Z-Statistic: -1.8831
# P-Value: 0.9702
# No significant difference in conversion rates (fail to reject null hypothesis).
```

```python
# 12.A tutoring service claims that its program improves students' exam scores. A sample of students who 
# participated in the program was taken, and their scores before and after the program were recorded.
# Use the below code to generate samples of respective arrays of marks:
# python
# before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
# after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])
# Use z-test to find if the claims made by tutor are true or false.

import numpy as np
from scipy.stats import norm

# Scores before and after the program
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate the differences
differences = after_program - before_program

# Calculate mean and standard deviation of differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the Z-statistic
z_stat = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # Two-tailed test

# Output results
print(f"Mean Difference: {mean_diff:.2f}")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("The tutoring program significantly improves exam scores (reject null hypothesis).")
else:
    print("No significant improvement in exam scores (fail to reject null hypothesis).")

# # ouput:
# Mean Difference: 3.80
# Z-Statistic: 4.5932
# P-Value: 0.0000
# The tutoring program significantly improves exam scores (reject null hypothesis).
```

```python
# 13.A pharmaceutical company wants to determine if a new drug is effective in reducing blood pressure. They 
# conduct a study and record blood pressure measurements before and after administering the drug.
# Use the below code to generate samples of respective arrays of blood pressure:
# python
# before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
# after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])
# Implement z-test to find if the drug really works or not.

import numpy as np
from scipy.stats import norm

# Blood pressure before and after the drug
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the differences
differences = before_drug - after_drug

# Calculate mean and standard deviation of differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Sample standard deviation
n = len(differences)

# Calculate the Z-statistic
z_stat = mean_diff / (std_diff / np.sqrt(n))

# Calculate the p-value
p_value = norm.cdf(z_stat)  # One-tailed test (drug reduces blood pressure)

# Output results
print(f"Mean Difference: {mean_diff:.2f}")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("The drug significantly reduces blood pressure (reject null hypothesis).")
else:
    print("No significant reduction in blood pressure (fail to reject null hypothesis).")

# output:
# Mean Difference: 10.10
# Z-Statistic: 10.0499
# P-Value: 1.0000
# No significant reduction in blood pressure (fail to reject null hypothesis).
```

```python
# 14.A customer service department claims that their average response time is less than 5 minutes. A sample 
# of recent customer interactions was taken, and the response times were recorded.
# Implement the below code to generate the array of response time:
# python
# response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])
# Implement z-test to find the claims made by customer service department are tru or false.

import numpy as np
from scipy.stats import norm

# Sample response times
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

# Hypothesized mean (claimed average response time)
mu = 5  # Population mean under null hypothesis

# Calculate sample mean and standard deviation
sample_mean = np.mean(response_times)
sample_std = np.std(response_times, ddof=1)  # Sample standard deviation
n = len(response_times)

# Calculate the Z-statistic
z_stat = (sample_mean - mu) / (sample_std / np.sqrt(n))

# Calculate the p-value for a one-tailed test
p_value = norm.cdf(z_stat)

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("The customer service department's claim is supported (reject null hypothesis).")
else:
    print("The customer service department's claim is not supported (fail to reject null hypothesis).")

# output:
# Sample Mean: 4.57
# Z-Statistic: -3.1845
# P-Value: 0.0007
# The customer service department's claim is supported (reject null hypothesis).
```

```python
# 15.A company is testing two different website layouts to see which one leads to higher click-through rates. 
# Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of 
# freedom, and p-value.
# Use the following data:
# python
# layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
# layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

import numpy as np
from scipy.stats import t

def ab_test(layout_a_clicks, layout_b_clicks, alpha=0.05):
    # Calculate the sample means and standard deviations
    mean_a = np.mean(layout_a_clicks)
    mean_b = np.mean(layout_b_clicks)
    std_a = np.std(layout_a_clicks, ddof=1)
    std_b = np.std(layout_b_clicks, ddof=1)
    
    # Calculate sample sizes
    n_a = len(layout_a_clicks)
    n_b = len(layout_b_clicks)
    
    # Calculate the t-statistic
    pooled_std = np.sqrt(((std_a**2) / n_a) + ((std_b**2) / n_b))
    t_stat = (mean_b - mean_a) / pooled_std
    
    # Calculate degrees of freedom using the Welch-Satterthwaite equation
    df = (((std_a**2) / n_a) + ((std_b**2) / n_b))**2 / (
        ((std_a**2 / n_a)**2 / (n_a - 1)) + ((std_b**2 / n_b)**2 / (n_b - 1))
    )
    
    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - t.cdf(abs(t_stat), df))
    
    # Print results
    print(f"Layout A Mean: {mean_a:.2f}, Standard Deviation: {std_a:.2f}")
    print(f"Layout B Mean: {mean_b:.2f}, Standard Deviation: {std_b:.2f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df:.4f}")
    print(f"P-Value: {p_value:.4f}")
    
    # Decision rule
    if p_value < alpha:
        print("There is a significant difference between the layouts (reject null hypothesis).")
    else:
        print("No significant difference between the layouts (fail to reject null hypothesis).")

# Test Data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform the A/B test
ab_test(layout_a_clicks, layout_b_clicks)

# output:
# Layout A Mean: 32.50, Standard Deviation: 3.03
# Layout B Mean: 42.00, Standard Deviation: 2.79
# T-Statistic: 7.2981
# Degrees of Freedom: 17.8799
# P-Value: 0.0000
# There is a significant difference between the layouts (reject null hypothesis).
```

```python
# 16.A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in 
# reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the t-
# statistic and p-value for the treatment effect.
# Use the following data of cholestrol level:
# python
# existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
# new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

import numpy as np
from scipy.stats import t

def compare_drugs(existing_drug_levels, new_drug_levels, alpha=0.05):
    # Calculate means and standard deviations
    mean_existing = np.mean(existing_drug_levels)
    mean_new = np.mean(new_drug_levels)
    std_existing = np.std(existing_drug_levels, ddof=1)
    std_new = np.std(new_drug_levels, ddof=1)
    
    # Calculate sample sizes
    n_existing = len(existing_drug_levels)
    n_new = len(new_drug_levels)
    
    # Calculate t-statistic
    pooled_std = np.sqrt(((std_existing**2) / n_existing) + ((std_new**2) / n_new))
    t_stat = (mean_existing - mean_new) / pooled_std
    
    # Calculate degrees of freedom using Welch-Satterthwaite equation
    df = (((std_existing**2) / n_existing) + ((std_new**2) / n_new))**2 / (
        ((std_existing**2 / n_existing)**2 / (n_existing - 1)) + ((std_new**2 / n_new)**2 / (n_new - 1))
    )
    
    # Calculate the p-value (one-tailed test: new drug < existing drug)
    p_value = t.cdf(-t_stat, df)
    
    # Print results
    print(f"Mean Cholesterol Level (Existing Drug): {mean_existing:.2f}")
    print(f"Mean Cholesterol Level (New Drug): {mean_new:.2f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df:.4f}")
    print(f"P-Value: {p_value:.4f}")
    
    # Decision rule
    if p_value < alpha:
        print("The new drug is significantly more effective than the existing drug (reject null hypothesis).")
    else:
        print("No significant difference between the drugs (fail to reject null hypothesis).")

# Cholesterol level data
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform the comparison
compare_drugs(existing_drug_levels, new_drug_levels)

# output:
# Mean Cholesterol Level (Existing Drug): 179.40
# Mean Cholesterol Level (New Drug): 171.90
# T-Statistic: 4.1405
# Degrees of Freedom: 17.8668
# P-Value: 0.0003
# The new drug is significantly more effective than the existing drug (reject null hypothesis).
```

```python
# 17.A school district introduces an educational intervention program to improve math scores. Write a Python 
# function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to 
# determine if the intervention had a significant impact.
# Use the following data of test score:
# python
# pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
# post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

import numpy as np
from scipy.stats import t

def analyze_intervention(pre_scores, post_scores, alpha=0.05):
    # Calculate the differences between post and pre scores
    differences = np.array(post_scores) - np.array(pre_scores)
    
    # Calculate the mean and standard deviation of the differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)
    
    # Calculate the sample size
    n = len(differences)
    
    # Calculate the t-statistic
    t_stat = mean_diff / (std_diff / np.sqrt(n))
    
    # Degrees of freedom (n - 1)
    df = n - 1
    
    # Calculate the p-value for a one-tailed test (assuming the intervention is expected to increase the scores)
    p_value = 1 - t.cdf(t_stat, df)
    
    # Print results
    print(f"Mean Difference: {mean_diff:.2f}")
    print(f"Standard Deviation of Differences: {std_diff:.2f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-Value: {p_value:.4f}")
    
    # Decision rule
    if p_value < alpha:
        print("The intervention had a significant impact (reject null hypothesis).")
    else:
        print("No significant impact from the intervention (fail to reject null hypothesis).")

# Test score data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform the analysis
analyze_intervention(pre_intervention_scores, post_intervention_scores)

# output:
# Mean Difference: 7.70
# Standard Deviation of Differences: 5.50
# T-Statistic: 4.4284
# Degrees of Freedom: 9
# P-Value: 0.0008
# The intervention had a significant impact (reject null hypothesis).
```

```python
# 18.An HR department wants to investigate if there's a gender-based salary gap within the company. Develop 
# a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically 
# significant difference between the average salaries of male and female employees.
# Use the below code to generate synthetic data:
# python
# # Generate synthetic salary data for male and female employees
# np.random.seed(0)  # For reproducibility
# male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
# female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

import numpy as np
from scipy.stats import t

def analyze_salary_gap(male_salaries, female_salaries, alpha=0.05):
    # Calculate means and standard deviations for both groups
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)
    std_male = np.std(male_salaries, ddof=1)
    std_female = np.std(female_salaries, ddof=1)
    
    # Calculate sample sizes
    n_male = len(male_salaries)
    n_female = len(female_salaries)
    
    # Calculate t-statistic
    pooled_std = np.sqrt(((std_male**2) / n_male) + ((std_female**2) / n_female))
    t_stat = (mean_male - mean_female) / pooled_std
    
    # Calculate degrees of freedom using Welch-Satterthwaite equation
    df = (((std_male**2) / n_male) + ((std_female**2) / n_female))**2 / (
        ((std_male**2 / n_male)**2 / (n_male - 1)) + ((std_female**2 / n_female)**2 / (n_female - 1))
    )
    
    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - t.cdf(np.abs(t_stat), df))
    
    # Print results
    print(f"Mean Salary (Male): {mean_male:.2f}")
    print(f"Mean Salary (Female): {mean_female:.2f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df:.4f}")
    print(f"P-Value: {p_value:.4f}")
    
    # Decision rule
    if p_value < alpha:
        print("There is a statistically significant gender-based salary gap (reject null hypothesis).")
    else:
        print("No statistically significant gender-based salary gap (fail to reject null hypothesis).")

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Perform the analysis
analyze_salary_gap(male_salaries, female_salaries)

# output:
# Mean Salary (Male): 55693.35
# Mean Salary (Female): 55501.75
# T-Statistic: 0.0611
# Degrees of Freedom: 36.1668
# P-Value: 0.9516
# No statistically significant gender-based salary gap (fail to reject null hypothesis).
```

```python
# 22. A company conducted a customer satisfaction survey to determine if there is a significant relationship 
# between product satisfaction levels (Satisfied, Neutral, Dissatisfied) and the region where customers are 
# located (East, West, North, South). The survey data is summarized in a contingency table. Conduct a Chi-
# Square test to determine if there is a significant relationship between product satisfaction levels and 
# customer regions.
# Sample data:
# python
#Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)
# data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])

import numpy as np
from scipy.stats import chi2_contingency

# Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)
data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Output the results
print("Contingency Table (Observed):")
print(data)
print("\nChi-Square Statistic:", chi2_stat)
print("P-Value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)

# Decision rule
alpha = 0.05
if p_value < alpha:
    print("\nThere is a significant relationship between product satisfaction levels and customer regions (reject null hypothesis).")
else:
    print("\nThere is no significant relationship between product satisfaction levels and customer regions (fail to reject null hypothesis).")

# output:
# Contingency Table (Observed):
# [[50 30 40 20]
#  [30 40 30 50]
#  [20 30 40 30]]

# Chi-Square Statistic: 27.777056277056275
# P-Value: 0.00010349448486004387
# Degrees of Freedom: 6
# Expected Frequencies:
# [[34.14634146 34.14634146 37.56097561 34.14634146]
#  [36.58536585 36.58536585 40.24390244 36.58536585]
#  [29.26829268 29.26829268 32.19512195 29.26829268]]

# There is a significant relationship between product satisfaction levels and customer regions (reject null hypothesis).
```

```python
# 24.A company produces three different versions of a product: Standard, Premium, and Deluxe. The 
# company wants to determine if there is a significant difference in customer satisfaction scores among the 
# three product versions. They conducted a survey and collected customer satisfaction scores for each 
# version from a random sample of customers. Perform an ANOVA test to determine if there is a significant 
# difference in customer satisfaction scores.
# Use the following data:
# python
# Sample data: Customer satisfaction scores for each product version
# standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
# premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
# deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

import numpy as np
from scipy.stats import f_oneway

# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform one-way ANOVA
f_stat, p_value = f_oneway(standard_scores, premium_scores, deluxe_scores)

# Output the results
print("F-Statistic:", f_stat)
print("P-Value:", p_value)

# Decision rule
alpha = 0.05
if p_value < alpha:
    print("\nThere is a significant difference in customer satisfaction scores among the product versions (reject null hypothesis).")
else:
    print("\nThere is no significant difference in customer satisfaction scores among the product versions (fail to reject null hypothesis).")

# output:
# F-Statistic: 27.03556231003039
# P-Value: 3.5786328857349003e-07
# There is a significant difference in customer satisfaction scores among the product versions (reject null hypothesis).
```
